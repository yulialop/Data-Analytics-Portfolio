## Import Packages and Data

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
import torch
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix, classification_report
# Set global font settings to Times New Roman, size 12
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = 12
```

```python
df = pd.read_csv("reviews_clean.csv")
```

```python
df.head()
```

## Emojis Analysis

```python
import emoji

# Function to detect if any emojis are present
def contains_emoji(text):
    return any(char in emoji.EMOJI_DATA for char in str(text))

# Apply to your DataFrame
df['contains_emoji'] = df['Review'].apply(contains_emoji)
```

```python
# Total number of reviews
total_reviews = len(df)

# Number of reviews that contain emojis
emoji_reviews = df['contains_emoji'].sum()

# Percentage
emoji_percentage = (emoji_reviews / total_reviews) * 100

# Display result
print(f"{emoji_percentage:.2f}% of reviews contain emojis.")
```

```python

# Function to remove emojis from text
def remove_emojis(text):
    return ''.join(char for char in str(text) if char not in emoji.EMOJI_DATA)

# Apply to remove emojis
df['Review'] = df['Review'].apply(remove_emojis)
df = df.drop(columns=['contains_emoji'])
```

## Sentiment Score

***

```python
tokenizer = AutoTokenizer.from_pretrained("nlptown/bert-base-multilingual-uncased-sentiment")

model = AutoModelForSequenceClassification.from_pretrained("nlptown/bert-base-multilingual-uncased-sentiment")
```

```python
def sentiment_score(text):
    tokens = tokenizer.encode(text, return_tensors='pt', truncation=True, padding=True)
    result = model(tokens)
    return int(torch.argmax(result.logits))+1
```

```python
# Apply to DataFrame
df['Sentiment Score'] = df['Review'].apply(sentiment_score)

# Add topics to reviews to see if the sentiment prediction is better
# df[['Review', 'main_category']].apply(Sentiment Score, axis=1)
```

Although both the Wilcoxon signed-rank test and the Chi-square test on the confusion matrix indicate a statistically significant difference between predicted and actual sentiment Ratings (p < 0.001), the mean absolute error (MAE) is 0.52, suggesting that the magnitude of mismatch is small in practical terms. This implies that the models predictions are typically within one star of the true Rating, which may be acceptable depending on the application. Therefore, while the model is not perfectly aligned with the proxy labels, its performance may still be considered strong from a business or user experience perspective.

```python
# Calculate off-by-one accuracy
off_by_one = np.abs(y_true - y_pred) <= 1
off_by_one_accuracy = np.mean(off_by_one)

print(f"Off-by-one accuracy: {off_by_one_accuracy:.4f}")
```

The sentiment classification model applied to user-generated reviews achieved an overall exact accuracy of 0.5620, meaning that approximately 56% of reviews were assigned the precise star Rating provided by the original reviewer. This performance aligns with the expected difficulty of the task, given the subjectivity of user Ratings and the inherent ambiguity of textual sentiment. The model demonstrated stronger performance for extreme Ratings (1-star and 5-star), as indicated by F1 scores of 0.64 and 0.73, respectively, while performance for intermediate Ratings (2-, 3-, and 4-star) was more modest.

Importantly, when allowing for a tolerance of 1 star, the off-by-one accuracy rises substantially to 0.9276. This means that in over 92% of cases, the models predicted sentiment was either exactly correct or differed from the user-provided rating by only a single star. This result is consistent with the performance reported by the developers of the nlptown/bert-base-multilingual-uncased-sentiment model, who observed 67% exact accuracy and 95% off-by-one accuracy on held-out English reviews. The high off-by-one accuracy underscores that most misclassifications occur between neighbouring ratings, as also reflected in the confusion matrix. Such small discrepancies are unsurprising, noting that user-generated star ratings, while serving as proxy sentiment labels, may not always represent consistent or objective sentiment. Such Ratings can be influenced by transient emotions, rating behaviour biases, or factors unrelated to textual sentiment (e.g., delivery issues, prior expectations). This introduces inherent noise into the ground truth labels, which may constrain the upper bound of model performance.

In summary, while the exact prediction of star ratings remains challenging due to label noise and human subjectivity, the model demonstrates a strong capability in producing sentiment ratings that are broadly aligned with human judgment when a margin of tolerance is considered.

### Analysis of Mismatches Between Ratings and Sentiment Scores in Customer Reviews

The observed discrepancies between low numerical Ratings and high positive sentiment scores primarily arise from several distinct factors:

1.	Updated or Evolving Opinions: Some reviewers initially expressed positive sentiment toward the product but subsequently updated their opinions due to price increases, freshness issues, or changes in quality. These updates often retain positive language from earlier experiences, yet the overall Rating reflects dissatisfaction (e.g., I love this coffee but its past its freshness date).
2.	Sarcasm and Irony: Several reviews employ sarcastic or ironic language that sentiment analysis models may misinterpret as positive due to the presence of ostensibly positive words or expressions. For example, phrases like perfect, scientific blending of sugar and salt and best chocolate chunk cookies ever within a negative review may mislead algorithms.
3.	Mixed or Nuanced Feedback: Many reviews contain positive comments about certain product aspects alongside strong criticisms related to price, packaging, or specific ingredients (e.g., great spice but I ordered the wrong spice or complaints about dented cans despite liking the product). Sentiment scores tend to capture the positive lexical cues while the overall Rating reflects the negative balance.
4.	Focus on External Factors Beyond Product Quality: Ratings often incorporate dissatisfaction with price, packaging, delivery conditions, or value for moneyelements not directly reflected in sentiment scores derived mainly from the review text content.
5.	Potential Model Limitations: Sentiment scoring models may lack domain-specific understanding, leading to misclassification when reviews contain complex sentiment, hedging, or contextual subtleties (e.g., contains GMO, so we will not buy again is likely scored positively due to neutral or positive wording around quality).
6.	Unexplained Rating Behavior: In a minority of cases, no clear rationale for Rating-sentiment incongruence emerges from the text, potentially indicating user error, Rating inconsistency, or extratextual influences.

The mismatch between low Ratings and high sentiment scores in these reviews can be attributed chiefly to evolving reviewer perspectives, sarcasm, nuanced or mixed sentiments, focus on extrinsic factors such as price and packaging, and inherent limitations of sentiment analysis models in capturing complex, context-dependent customer opinions.

### Analysis of High Ratings with Low Sentiment Scores in Customer Reviews

Instances where customers assign high numerical Ratings but exhibit low or negative sentiment scores in the text typically reflect the following patterns:

1. Reluctant Positive Ratings Despite Reservations: Reviewers often appreciate the overall product or brand and thus assign a high Rating but express dissatisfaction or concerns in the textual review. This includes complaints about price, quantity, packaging, or minor quality issues that lower the sentiment score but do not prevent them from Rating favourably.
2.	Polite or Diplomatic Criticism: Some reviewers frame their negative experiences in tempered or polite language, which may register as low sentiment by the algorithm but still warrant a relatively high Rating, reflecting satisfaction with the core product despite flaws.
3.	Emphasis on Functional Benefits Over Subjective Experience: Positive Ratings may reflect product usefulness or meeting expectations, while the review text highlights specific shortcomings (e.g., works great but the amount is too much to finish, or good flavour but too expensive), causing sentiment scores to appear negative.
4.	Textual Focus on External or Secondary Factors: Criticism centred on packaging damage, delivery issues, or pricing strategy can reduce sentiment scores even when the product itself is rated highly.
5.	Model Sensitivity to Negative Keywords: Sentiment models can overweight certain negative words or phrases (expensive, dented, dry, small quantity) which decrease sentiment scores, whereas customers overall satisfaction is captured by their high Rating.
6.	Possible Ambiguity or Mixed Emotions: The coexistence of positive and negative expressions within a review creates a complex emotional tone that may be misinterpreted by simplistic sentiment models, yielding a low score despite a high Rating.

High Ratinglow sentiment mismatches primarily stem from reviewers expressing nuanced critiques or external concerns within their text, while still endorsing the product overall. Such cases highlight the limitation of sentiment analysis tools in fully capturing complex, multifaceted consumer opinions that balance appreciation with constructive feedback.

### Analysis of Rating and Sentiment Score Mismatches in Customer Reviews

Discrepancies between numerical Ratings and textual sentiment scores arise from several key factors:

- Evolving Opinions and Updated Reviews: Customers may retain positive language in their text from earlier experiences but adjust their Ratings downward due to changes in product quality, price, or freshness.
- Sarcasm, Irony, and Nuanced Language: Sentiment models often misinterpret sarcastic or ironic remarks as positive, while the Rating reflects genuine dissatisfaction. Similarly, mixed sentiments within reviews can confuse models that focus on lexical cues.
- Focus on Extrinsic Factors: Ratings frequently incorporate judgments on price, packaging, delivery, and valueelements that sentiment scores derived from review text alone may not capture accurately.
- Polite or Diplomatic Criticism: Some high Ratings coexist with low sentiment when reviewers acknowledge product benefits but criticize aspects like cost or quantity in a subdued manner.
- Model Limitations in Context Understanding: Automated sentiment analysis struggles with complex emotional tones, domain-specific nuances, and implicit meanings, resulting in mismatches.
- Unexplained or Inconsistent Behavior: Occasionally, Rating-text incongruences may reflect user Rating errors or extratextual influences not evident in the review content.

These patterns highlight inherent challenges in aligning numerical Ratings with sentiment analysis and underscore the need for more sophisticated, context-aware approaches when interpreting customer feedback.

# Emotion Extraction

***

For the purpose of emotion extraction in customer review analysis, the model j-hartmann/emotion-english-distilroberta-base model generates continuous probability scores for Ekmans six basic emotions (Ekman, 1992), labelled by the model as joy (which Ekman refers to as enjoyment), anger, fear, disgust, sadness, and surprise, plus a neutral category. Each reviews emotional profile is represented by probabilities that sum to 1 across the seven categories. This probabilistic output captures the full range of variation in emotional signals, offering more nuanced insights and improving predictive and clustering performance
(Dawson and Weiss, 2012).

```python
# Set up the emotion classifier
classifier = pipeline(
    "text-classification",
    model="j-hartmann/emotion-english-distilroberta-base",
    top_k=None,
    truncation=True
)

# Prepare to store results
emotion_vectors = []
dominant_emotions = []

# Iterate through reviews
for review in df["Review"]:
    result = classifier(review)[0]  # list of dicts with 'label' and 'score'
    
    # Create a dict of label: score
    score_dict = {r["label"].capitalize(): r["score"] for r in result}
    
    # Store full score vector
    emotion_vectors.append(score_dict)
    
    # Store top emotion
    top_emotion = max(result, key=lambda x: x["score"])["label"].capitalize()
    dominant_emotions.append(top_emotion)

# Convert score vectors into a DataFrame
emotion_scores_df = pd.DataFrame(emotion_vectors)

# Add the dominant emotion
emotion_scores_df["Dominant Emotion"] = dominant_emotions

# Concatenate with the original DataFrame
df = pd.concat([df, emotion_scores_df], axis=1)
```

### Statistical Significance
A chi-squared test of independence was conducted to assess the association between predicted dominant emotion and sentiment score (on a 15 scale). The result was highly significant, indicating that the distribution of dominant emotions varies systematically across different levels of sentiment. The effect size, measured using Cramrs V, suggests a relatively strong association between the two variables.

This finding implies that the models predicted emotional labels are not randomly distributed with respect to sentiment intensity. Instead, certain emotions are more likely to occur in reviews with higher or lower sentiment scores. This supports the construct validity of the emotion classification output and confirms its alignment with more traditional sentiment measures, offering a more nuanced emotional interpretation of customer feedback.



```python
from scipy.stats import chi2_contingency

contingency = pd.crosstab(df["Dominant Emotion"], df["Sentiment Score"])
chi2, p, dof, expected = chi2_contingency(contingency)

print(f"Chi2 = {chi2}, p-value = {p}")
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = stats.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))  # bias correction
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

cramers_v(df["Dominant Emotion"], df["Sentiment Score"])
```

### Zero-Shot Classification

***

```python
from transformers import pipeline
import pandas as pd
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Combine all reviews into one text
text = " ".join(df['Review'].dropna().astype(str))

# Generate the word cloud
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    colormap='Blues'  # try 'Set2', 'cool', 'Dark2', etc.
).generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
# Your candidate labels
candidate_labels = ['Packaging', 'Price', 'Value for Money', 'Taste and Flavour', 'Freshness', 'Expiration Date', 'Delivery', 'Ingredients', 'Quality', 'Size and Quantity', 'Easy to Use', 'Not as Expected', 'Returns', 'Damaged']
# Load the zero-shot classifier
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")


# Function to classify a single piece of text and return a dict of scores
def classify_review(text):
    result = classifier(text, candidate_labels, multi_label=True)
    return {label: score for label, score in zip(result['labels'], result['scores'])}

# Apply the classification to each review
scores_df = df['Review'].apply(classify_review).apply(pd.Series)

# Concatenate the original df with the scores
df = pd.concat([df, scores_df], axis=1)

# Find dominant topic and its score
dominant_topic = scores_df.idxmax(axis=1)
dominant_score = scores_df.max(axis=1)

# Apply threshold: if score < threshold, assign 'No dominant topic'
df['Dominant Topic'] = dominant_topic.where(dominant_score >= 0.5, 'Other')
```
